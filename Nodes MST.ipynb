```python
# for some basic operations
import numpy as np 
import pandas as pd 

# for basic visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# for network visualizations
import networkx as nx
```

```python
edgelist = [['Node1', 'Node2', 85], 
            ['Node2', 'Node3', 80], 
            ['Node3', 'Node1', 186], 
            ['Node3', 'Node4', 167], 
            ['Node4', 'Node5', 84], 
            ['Node4', 'Node6', 502], 
            ['Node4', 'Node7', 183], 
            ['Node6', 'Node8', 103], 
            ['Node6', 'Node9', 167],
            ['Node3', 'Node8', 1000],
            ['Node3', 'Node6', 167],
            ['Node7', 'Node9', 209],
            ['Node5', 'Node2', 76],
            ['Node5', 'Node7', 154],
            ['Node1', 'Node8', 121]]
```

```python
# Undirected Graphs
'''
['bipartite_layout',
 'circular_layout',
 'kamada_kawai_layout',
 'random_layout',
 'rescale_layout',
 'shell_layout',
 'spring_layout',
 'spectral_layout',
 'fruchterman_reingold_layout']
 '''
g = nx.Graph()
for edge in edgelist:
    g.add_edge(edge[0],edge[1], weight = edge[2])

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (11, 11)
plt.style.use('fivethirtyeight')

pos = nx.spring_layout(g)

# drawing nodes
nx.draw_networkx(g,pos)
#plt.title('Undirected Graphs', fontsize = 20)
plt.axis('off')
plt.show()
```
